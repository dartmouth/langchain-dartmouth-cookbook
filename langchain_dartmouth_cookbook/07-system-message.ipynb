# System message

In addition to messages from the human and the AI, there is a third kind of message called _system message_. The system message is a way to give context to the conversation and provide the model with instructions that underpin the entire conversation, such as a particular persona. Let's explore this!

```python
from dotenv import find_dotenv, load_dotenv

load_dotenv(find_dotenv())
```

## Persona

Without a system message, the model goes with its default "helpful assistant"-style persona:

```python
from langchain_dartmouth.llms import ChatDartmouth

llm = ChatDartmouth(model_name="llama-3-1-8b-instruct")
response = llm.invoke("I have a lot of stuff to do and don't know where to start!")
response.pretty_print()
```

We can override the persona and give extra context by providing a system message as the first message in a conversation:

```python
conversation = [
    (
        "system",
        "You are the children's television host Mr. Rogers talking to an 8 year old child.",
    )
]
conversation.append(
    ("human", "I have a lot of stuff to do and don't know where to start!")
)

response = llm.invoke(conversation)
response.pretty_print()
```

Notice how the model is adopting the described persona, which not only affects the tone of the response but also the level of complexity at which the contents are presented. 

This is one useful application of the system message: Control multiple dimensions of the response using a persona.


## Guardrailing

Another important use-case for the system message is "guardrailing" to control what kind of requests the model will refuse to comply with. For example, we can provide explicit instructions to _not_ engange in certain behavior:

```python
conversation = [
    (
        "system",
        "You are an assistant for meal plan generation. Do not engange in any conversation outside of this scope.",
    ),
    (
        "human",
        "Tell me a joke!",
    ),
]

response = llm.invoke(conversation)
response.pretty_print()
```

## Formatting and capabilities
Generally speaking, any instructions we want to apply to every response can go into the system message. Another example for this would be the desired output formatting:

```python
conversation = [
    (
        "system",
        "Always respond in valid JSON format.",
    ),
    (
        "human",
        "Tell me a joke!",
    ),
]

response = llm.invoke(conversation)
response.pretty_print()
```

System messages can also be used to describe capabilities and strategies to the model:

```python
conversation = [
    (
        "system",
        "You are a meal planner. You can provide a weekly meal plan for your client. Gather all necessary information before generating the plan.",
    ),
    (
        "human",
        "Hi! Can you help me?",
    ),
]

response = llm.invoke(conversation)
response.pretty_print()
```

The system message can even include a description of certain tools the model may be able to use. We will learn more about that in [the recipe on tool use](13-tool-use.ipynb)!

## Summary

System messages are a powerful tool to ground the conversation in a particular context, give the model global instructions, imbue it with a persona, and control the scope of the conversation.
